5. Usando una figura Y EJES POR SEPARADO, usa numpy para los valores del eje X entre 0 y 5. Pinta dos gráficas en dos cajas distintas, a la izquierda una recta con pendiente positiva de 3 que pase por (0,0) y a la derecha una recta con pendiente negativa de 3 que pase por (0,-5). Elige la precisión en el eje X que desees.  

```python

x = np.linspace(0, 10, 100)  #Crea un array de 100 puntos espaciados uniformemente entre 0 y 5 usando NumPy. 
#SU UUSO ES Para definir ejes en gráficos (por ejemplo, plt.plot(x, y))
fig, axes =plt.subplots(1,2, figsize=(10,4))



x1 = 3*x  #recta con pendiente psitiva (paa por 0,0)
x2 = -3 * x - 5 # ecta cn pendiente negativa   -3 puntos, pasando por  0, -5
#AHORA HACEMOS LA FIGURA CON DOS SUBGRAFICAS UNA AL LADO DE OTRA

axes[0].plot(x,x1)
axes[0].set_xlim([0, 4])  # Establecer límites en el eje X
axes[0].set_ylim(-15, 15) #establecer límites en el eje y
axes[0].set_xlabel("eje X")
axes[0].set_ylabel("eje Y")
axes[0].set_title("Recta Positiva")


axes[1].plot(x,x2)
axes[1].set_xlim([0, 4])  # Establecer límites en el eje X
axes[1].set_ylim(-15, 15)
axes[1].set_xlabel("eje X")
axes[1].set_ylabel("eje Y")
axes[1].set_title("Recta Negativa")



# una forma de hacerlo más limpio y en una sla linea 

axes[0].plot(x, x1)
axes[1].plot(x, x2)

# Establecer atributos comunes de los ejes en una sola línea
for ax in axes:
    ax.set(xlabel='Eje X', ylabel='Eje Y', xlim=(0, 4), ylim=(-15, 15))
```

```python
x = np.linspace(0, 5)  #Crea un array de 100 puntos espaciados uniformemente entre 0 y 5 usando NumPy. 
#SU UUSO ES Para definir ejes en gráficos (por ejemplo, plt.plot(x, y))

x1 = 3*x  #recta con pendiente psitiva (paa por 0,0)
x2 = -3 * x - 5 # ecta cn pendiente negativa   -3 puntos, pasando por  0, -5
#AHORA HACEMOS LA FIGURA CON DOS SUBGRAFICAS UNA AL LADO DE OTRA

#gráfico de la izquierda 
#plt.figure(figsize=(10, 5))  #tamañ de la figura 
plt.subplot(1, 2, 1)  # 1 fila, 2 columnas, posición 1
plt.plot(x, x1) #graficamos la recta positiva
plt.title("Recta positiva")


#gráfico de la derecha
#plt.figure(figsize=(10, 5))  #tamañ de la figura 
plt.subplot(1, 2, 2)  # 1 fila, 2 columnas, posición 1
plt.plot(x, x2) #graficamos la recta positiva
plt.title("Recta negativa")
# Ajustar el espacio entre subgráficas
plt.tight_layout()

```

19. Vamos a pintar un histograma de unas alturas generadas aleatoriamente. Crea 300 alturas entre 150 y 190 cm.

```python
# Generar 300 alturas aleatorias entre 150 y 190 cm
altura = np.random.randint(150, 191, size=30000)

# Crear el histograma
plt.hist(altura, bins=10, color='blue', edgecolor='black')

# Títulos y etiquetas
plt.title('Histograma de Alturas (150 - 190 cm)', fontsize=14)
plt.ylabel('Altura (cm)')
plt.xlabel('Frecuencia')
```
